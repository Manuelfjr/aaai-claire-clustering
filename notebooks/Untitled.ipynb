```python
import concurrent.futures
import math
import time

class ParallelProcessor:
    def __init__(self, max_workers=None):
        self.max_workers = max_workers

    def process_parallel(self, numbers):
        with concurrent.futures.ThreadPoolExecutor(max_workers=self.max_workers) as executor:
            results = list(executor.map(math.factorial, numbers))
        return results

def process_sequential(numbers):
    results = [math.factorial(number) for number in numbers]
    return results

# Generate a list of numbers for processing
numbers = list(range(1, 10000001))

# Create an instance of ParallelProcessor
pe = ParallelProcessor(16)

# Measure time taken for parallel processing
start_time_parallel = time.time()
parallel_results = pe.process_parallel(numbers)
end_time_parallel = time.time()

# Measure time taken for sequential processing
start_time_sequential = time.time()
sequential_results = process_sequential(numbers)
end_time_sequential = time.time()

# Compare the processing times
parallel_duration = end_time_parallel - start_time_parallel
sequential_duration = end_time_sequential - start_time_sequential

print(f"Parallel processing took {parallel_duration:.4f} seconds.")
print(f"Sequential processing took {sequential_duration:.4f} seconds.")
```
